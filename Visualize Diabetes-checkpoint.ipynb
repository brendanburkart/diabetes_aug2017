# visualize features

```python
plt.hist(data.patient_frequency, bins = 20, alpha=0.5, color = 'blue', label='ALL')

# FORMATING FOR PLOT
plt.legend(loc='upper right')
plt.title('Distribution of Patient Frequency', fontsize = 15)
plt.ylabel('Number of Patients', fontsize = 15)
plt.xlabel('Patient Frequency', fontsize = 15)

plt.show()
```

```python
# PLOT PATIENT FREQUENCY

plt.hist(data.patient_frequency[(data.patient_frequency < 15)], \
         bins = 10, alpha=0.2, color = 'blue', label='ALL')

plt.hist(data.patient_frequency[(data.patient_frequency < 15) & (data.readmitted == True)], \
         bins = 10, alpha=0.5, color = 'red', label='READMITTED')

plt.hist(data.patient_frequency[(data.patient_frequency < 15) & (data.readmitted == False)], \
         bins = 10, alpha=0.4, color = 'green', label='NOT readmitted')

# FORMATING FOR PLOT
plt.legend(loc='upper right')
plt.title('Distribution of Patient Frequency (where freq < 15)', fontsize = 15)
plt.ylabel('Number of Patients', fontsize = 15)
plt.xlabel('Patient Frequency', fontsize = 15)

plt.show()
```

```python
# PLOT GENDER

data.gender[data.readmitted == False].value_counts().plot(kind='bar', color = 'green', alpha = 0.4)
data.gender[data.readmitted == True].value_counts().plot(kind='bar', color = 'red', alpha = 0.4)
```

```python
# PLOT RACE

data.race[data.readmitted == False].value_counts().plot(kind='bar', color = 'green', alpha = 0.4)
data.race[data.readmitted == True].value_counts().plot(kind='bar', color = 'red', alpha = 0.4)
```

```python
# PLOT AGE
print('Patient age known for ', data.age[data.age != 0].count(), ' patients')
print()

plt.hist(data.age[data.readmitted == False], \
         bins = 10, alpha=0.5, color = 'green', label='NOT readmitted')
plt.hist(data.age[data.readmitted == True], \
         bins = 10, alpha=0.5, color = 'red', label='READMITTED')

# FORMATING FOR PLOT
plt.legend(loc='upper left')
plt.title('Distribution of Patient Age, 100% known', fontsize = 15)
plt.ylabel('Number of Patients', fontsize = 15)
plt.xlabel('Patient Age', fontsize = 15)

plt.show()
```

```python
# PLOT WEIGHT (minus the unknowns)
plt.hist(data.weight[(data.weight != 0) & (data.readmitted == True)], \
         bins = 12, alpha=0.5, color = 'red', label='READMITTED')
plt.hist(data.weight[(data.weight != 0) & (data.readmitted == False)], \
         bins = 12, alpha=0.5, color = 'green', label='NOT readmitted')

# FORMATING FOR PLOT
plt.legend(loc='upper left')
plt.title('Distribution of the 3,197 Known Patient Weights (3.14% of the total)', fontsize = 15)
plt.ylabel('Number of Patients', fontsize = 15)
plt.xlabel('Patient Weight', fontsize = 15)

plt.show()
```

```python
# PLOT 30 MOST COMMON DIAGNOSES

# arranged by most common diagnosis, and overlayed to see readmitted category

data.diag_1_stem.value_counts().head(30)\
    .plot(kind='bar', color = 'blue', alpha = 0.1, label = 'ALL')
    
data.diag_1_stem[data.readmitted == True].value_counts().head(30)\
    .plot(kind='bar', color = 'red', alpha = 0.4, label = 'READMITTED')
    
data.diag_1_stem[data.readmitted == False].value_counts().head(30)\
    .plot(kind='bar', color = 'green', alpha = 0.4, label = 'Not')

# FORMATING FOR PLOT
plt.legend(loc='upper right')
plt.title('Top 30 Primary Diagnoses', fontsize = 15)
plt.ylabel('Number of Patients', fontsize = 15)
plt.xlabel('Diagnosis ICD-9', fontsize = 15)
```

```python
# PLOT 200 LEAST COMMON DIAGNOSES

# arranged by most common diagnosis, and overlayed to see readmitted category

data.diag_1_stem.value_counts().tail(200)\
    .plot(kind='bar', color = 'blue', alpha = 0.1, label = 'ALL')
    
data.diag_1_stem[data.readmitted == True].value_counts().tail(200)\
    .plot(kind='bar', color = 'red', alpha = 0.4, label = 'READMITTED')
    
data.diag_1_stem[data.readmitted == False].value_counts().tail(200)\
    .plot(kind='bar', color = 'green', alpha = 0.4, label = 'Not')

# FORMATING FOR PLOT
plt.legend(loc='upper right')
plt.title('200 Least Common Diagnoses', fontsize = 15)
plt.ylabel('Number of Patients', fontsize = 15)
plt.xlabel('Diagnosis ICD-9', fontsize = 15)
```
